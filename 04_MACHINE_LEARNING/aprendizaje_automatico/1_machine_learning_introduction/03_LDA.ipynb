# Linear discriminant analysis (LDA)

In this notebook we will learn how to apply LDA to a given dataset using the <a href="https://scikit-learn.org/stable/" target="_blank">scikit-learn</a> library. 

Linear Discriminant Analysis (LDA) tries to identify the features that account for the most variance between classes. In particular, LDA, in contrast to PCA, is a supervised method, using known class labels.

First we load the Iris dataset.

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names
```

Now we create an instance of the <a href="https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html"><code>sklearn.discriminant_analysis.LinearDiscriminantAnalysis</code></a> class, indicating the desired dimensionality reduction with <code>n_components=2</code>, and fit the model to our data with the <code>fit()</code> method. 

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X,y)
```

The fitted model can now be used to reduce the dimensionality of the input data by projecting it to the most discriminative directions, using the <code>transform()</code> method. Remember that the original data has 4 components, i.e. the features' data matrix is an ndarray of shape <code>(150, 4)</code>.

```python
X_transformed = lda.transform(X)


# display the projected dataset
import matplotlib.pyplot as plt
for i, target_name in zip([0, 1, 2], target_names):
    plt.scatter(X_transformed[y == i, 0], X_transformed[y == i, 1], label=target_name)
plt.legend()
```
